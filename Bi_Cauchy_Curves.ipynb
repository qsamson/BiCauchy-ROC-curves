```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.title("Bi-Cauchy ROC Curves")
pi = math.pi
tan = np.tan
arctan = np.arctan
x = np.linspace(0.00001, 1, num=100)
b = [0.5,1.5,2.5,3.5]
a = [1,2,3,4]
colors = ['purple','brown','pink', 'olive','cyan','red','green','indigo','darkblue','darkcyan']
col_count = 0
h = []
for i in range(len(a)):
  y= (1/2) + (1/pi)*arctan(a[i] + b[i]*tan(pi*x +  (pi/2)))
  globals()[f"h_{col_count}"], = plt.plot(x,y,color = colors[col_count], label="\u03B1 = "+str(a[i])+", \u03B2 = " + str(b[i]))
  h.append(globals()[f"h_{col_count}"],)
  col_count = col_count + 1
leg = plt.legend(handles=h, loc='lower right')
ax = plt.gca().add_artist(leg)

fig = plt.gcf()
plt.grid()
plt.show()
```
